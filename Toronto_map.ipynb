```python
import pandas as pd
import numpy as np
```

```python
dfToronto1 = pd.read_excel('C:/Users/LuciaBenisova/Toronto1.xlsx')
dfToronto1.head()
```

```python
dfToronto1['TorontoFlag'] = np.where(dfToronto1['Borough'].str.contains("Toronto", case=False, na=False), 'Y', '')
dfToronto1
```

```python
from geopy.geocoders import Nominatim
```

```python
toronto_data = dfToronto1[dfToronto1['TorontoFlag'] == 'Y'].reset_index(drop=True)
toronto_data.head()
```

```python
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
import folium
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neigborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
